```python
from async_savers import load_shards
import os
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  
import matplotlib.pyplot as plt

from franka_robot import FrankaRobot
```

```python
fr = FrankaRobot()
```

```python
qs0 = np.array(load_shards('sampled_qs_0'))
ps0 = np.array([fr.ee(q)[:3] for q in qs0])

qs1 = np.array(load_shards('sampled_qs_1'))
ps1 = np.array([fr.ee(q)[:3] for q in qs1])
```

```python
dists = np.array(load_shards('sampled_qs_dist'))
```

```python
%matplotlib inline
plt.plot(np.arange(len(dists)), dists)
plt.show()
```

```python
%matplotlib notebook

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 600

ax.scatter(ps0[:n,0], ps0[:n,1], ps0[:n,2], c='b')
ax.scatter(ps1[:n,0], ps1[:n,1], ps1[:n,2], c='r')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
